----
## $Electrostatico$
- $Energia\ almacenada\ 3.42677\cdot 10^{-10}J\ para\ cuarto\ simetria$
----

```python
conf6Electrostatico = pd.read_csv("Configuracion 6\Electrostatico\Media simetria\Configuracion 6.txt", sep='\t')
conf6Electrostatico
```

```python
fig = plt.figure(figsize=[16, 9])
ax = fig.add_subplot(1,1,1)
#plt.style.use("fast")
#Aca establezco la subdivicion del grafico
ax.set_xticks(np.linspace(0,160, 33), minor=True)
ax.set_yticks(np.linspace(0,600, 13), minor=True)

ax.tick_params(axis='both', labelsize=20)

#Parametros en los cuatro lados
plt.tick_params(axis='both', labelsize=20, labelleft=True, labelright=False, labeltop=False, labelbottom=True)


#1000*EJERCICIO_A_1_A['time'] es para poner la escala en mili segundos

#SEÑAL DE ENTRADA Vi
plt.plot(conf6Electrostatico['Length'],conf6Electrostatico['Field'],label='$Campo\ Eléctrico:\ E[V/m]$')


#EJE HORIZONTAL
ax.axhline(0,linestyle='-',linewidth=2,c='k')
ax.axvline(0,linestyle='-',linewidth=5,c='k')

#Limites del grafico
plt.xlim(0, 150)
plt.ylim(0, 600)

#Nombre de los ejes
plt.xlabel('$Distancia[mm]$', fontsize=25)
plt.ylabel('$Campo\ Eléctrico[V/m]$', fontsize=25)

#Parametros del recuadro con nombre de las curvas y tamaño de la fuente
plt.rcParams['figure.facecolor'] = 'white'  # bluish dark grey
plt.rcParams['axes.facecolor'] = 'white' # bluish dark grey
plt.rcParams['savefig.facecolor'] = 'white'  # bluish dark grey

for param in ['axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = 'k'  # very light grey

plt.rcParams['text.color']='k'
plt.grid(which='both', linestyle='--',color='grey')
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"


legend = ax.legend(loc=1, shadow=True, fontsize=18)
legend.get_frame().set_facecolor('0.9')
plt.savefig('Campo Electrico Configuracion 6.svg',format="svg",bbox_inches='tight')
```
